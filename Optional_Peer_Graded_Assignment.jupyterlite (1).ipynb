```python
##Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
## State the hypothesis:
## H_0:_1=_2 ("there is no difference in median value of houses bounded by the river")
## H_0:_1=/=_2 ("there is difference in median value of houses bounded by the river")

ax = sns.distplot(boston_df['MEDV'],
                  bins=20,
                  kde=True,
                  color='red',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1]['MEDV'], equal_var = True)
```

```python
## Since the p-value is less than alpha value 0.05,  there is a statistical difference in in median value of houses bounded by the river
```

```python
## Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)
## It is created 3 different groups:
## 35 years and younger, between 35 and 70 years and 70 years and older

boston_df.loc[(boston_df['AGE'] <= 33), 'age_group'] = '35 years and younger'
boston_df.loc[(boston_df['AGE'] > 33)&(boston_df['AGE'] < 66), 'age_group'] = 'between 35 and 70 year'
boston_df.loc[(boston_df['AGE'] >= 66), 'age_group'] = '70 years and older'

less_lower = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
between_higher = boston_df[boston_df['age_group'] == 'between 35 and 70 year']['MEDV']
more_higher = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']

## running ANOVA
f_statistic, p_value = scipy.stats.f_oneway(less_lower, between_higher,more_higher )
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

```python
##Since the p-value is less than 0.05, we will reject the null hypothesis as there is significant evidence that at least one of the means differ.
```
